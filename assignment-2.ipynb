```python
import pandas as pd
import sqlite3

# Load dataset
df = pd.read_csv(
    '/kaggle/input/datasets/vivek468/superstore-dataset-final/Sample - Superstore.csv',
    encoding='latin1'
)

# Create database
conn = sqlite3.connect('superstore.db')

# Load data into SQL table
df.to_sql('superstore', conn, if_exists='replace', index=False)

print("Dataset loaded successfully!")
```

```python
#View Schema
query = """
PRAGMA table_info(superstore);
"""

pd.read_sql(query, conn)
```

```python
#Sample Data
query = """
SELECT *
FROM superstore
LIMIT 10;
"""

pd.read_sql(query, conn)
```

```python
#Region Filter

query = """
SELECT *
FROM superstore
WHERE Region = 'West';
"""

pd.read_sql(query, conn)
```

```python
#Category Filter

query = """
SELECT *
FROM superstore
WHERE Category = 'Technology';
"""

pd.read_sql(query, conn)
```

```python
#Sales Filter

query = """
SELECT *
FROM superstore
WHERE Sales > 500;
"""

pd.read_sql(query, conn)
```

```python
#Data filter

query = """
SELECT *
FROM superstore
WHERE [Order Date] >= '2017-01-01';
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT COUNT(DISTINCT [Order ID]) AS Unique_Orders
FROM superstore;
"""

pd.read_sql(query, conn)
```

# ##GROUP BY Aggregations

```python
#Total Sales by Category

query = """
SELECT Category,
       SUM(Sales) AS Total_Sales
FROM superstore
GROUP BY Category;
"""

pd.read_sql(query, conn)
```

```python
#Total Quantity by Category

query = """
SELECT Category,
       SUM(Quantity) AS Total_Quantity
FROM superstore
GROUP BY Category;
"""

pd.read_sql(query, conn)
```

```python
#Average Sales by Region

query = """
SELECT Region,
       ROUND(AVG(Sales),2) AS Avg_Sales
FROM superstore
GROUP BY Region;
"""

pd.read_sql(query, conn)
```

# Sort and Limit Results

```python
#Top 10 Products

query = """
SELECT [Product Name],
       SUM(Sales) AS Total_Sales
FROM superstore
GROUP BY [Product Name]
ORDER BY Total_Sales DESC
LIMIT 10;
"""

pd.read_sql(query, conn)
```

```python
#Top Categories

query = """
SELECT Category,
       SUM(Sales) AS Total_Sales
FROM superstore
GROUP BY Category
ORDER BY Total_Sales DESC;
"""

pd.read_sql(query, conn)
```

# Business Use Cases

```python
#Monthly Sales Trend

query = """
SELECT substr([Order Date], 1, 7) AS Month,
       SUM(Sales) AS Total_Sales
FROM superstore
GROUP BY Month
ORDER BY Month;
"""

pd.read_sql(query, conn)
```

```python
#Top 10 Customers

query = """
SELECT [Customer Name],
       SUM(Sales) AS Total_Sales
FROM superstore
GROUP BY [Customer Name]
ORDER BY Total_Sales DESC
LIMIT 10;
"""

pd.read_sql(query, conn)
```

```python
#Duplicate Orders

query = """
SELECT [Order ID],
       COUNT(*) AS Duplicate_Count
FROM superstore
GROUP BY [Order ID]
HAVING COUNT(*) > 1;
"""

pd.read_sql(query, conn)
```

# Validation Queries

```python
#Row Count

query = """
SELECT COUNT(*) AS Total_Rows
FROM superstore;
"""

pd.read_sql(query, conn)
```

```python
#Null Values Check

query = """
SELECT *
FROM superstore
WHERE Sales IS NULL
   OR Quantity IS NULL
   OR Profit IS NULL;
"""

pd.read_sql(query, conn)
```
